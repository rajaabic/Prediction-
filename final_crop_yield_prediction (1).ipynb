```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
df=pd.read_csv("crop_yield.csv")
df
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
c=df["Crop"].nunique()
print(c)
s=df["Season"].nunique()
print(s)
s1=df["State"].nunique()
print(s1)
```

```python
df_encoded=pd.get_dummies(df,columns=["Crop","Season","State"],drop_first=True)
```

```python
df_encoded.isnull().sum()
```

```python
df_encoded.describe()
```

```python
x=df_encoded.drop('Yield',axis=1)
y=df_encoded['Yield']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y,test_size=0.2,random_state=42)
```

```python
numeric_cols=["Crop_Year","Area","Production","Annual_Rainfall","Fertilizer","Pesticide"]
```

```python
scaler=StandardScaler()
x[numeric_cols]=scaler.fit_transform(x[numeric_cols])
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred_lr=lr.predict(x_test)
print(y_pred_lr)
```

```python
dt=DecisionTreeRegressor(random_state=42)
dt.fit(x_train,y_train)
y_pred_dt=dt.predict(x_test)
print(y_pred_dt)
```

```python
rf= RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)
print(y_pred_rf)
```

```python
gb=GradientBoostingRegressor(random_state=42)
gb.fit(x_train,y_train)
y_pred_gb=gb.predict(x_test)
print(y_pred_gb)
```

```python
def evaluate(y_test,y_pred,name):
    print(f"\n{name}")
    print("MAE :",mean_absolute_error(y_test,y_pred))
    print("MSE:",mean_squared_error(y_test,y_pred))
    print("R2  :",r2_score(y_test,y_pred))
```

```python
evaluate(y_test,y_pred_lr,"Linear Regression")
evaluate(y_test,y_pred_dt,"Decision Tree")
evaluate(y_test,y_pred_rf,"Random Forest")
evaluate(y_test,y_pred_gb,"Gradient Boosting")
```
